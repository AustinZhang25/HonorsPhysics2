```python
while t<=1:
    stretch = startX-h
    fspring = k*stretch
    a = ((k*stretch) + (-m*g))/m
    v += a*dt
    h += v*dt
    t += dt
    GPE = m*g*h
    EPE = (k*stretch*stretch)/2
    KE = (m*v**2)/2
    time.append(t)
    List.append(h)
    ListGPE.append(GPE)
    ListEPE.append(EPE)
    ListKE.append(KE)
    NetE = GPE + EPE + KE
    ListNetE.append(NetE)
    plt.plot(time, List, color="blue")
    plt.plot(time, ListGPE, color="green")
    plt.plot(time, ListEPE, color="red")
    plt.plot(time, ListKE, color="purple")
plt.legend(["Position", "GPE", "EPE", "KE"])
plt.show()
```
