```python
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas as pd
```

```python
h = 0.5
startX = 0.7
stretch = startX-h
v = 0
k = 5
m = 0.0609
g = 9.8
w = (9.8*m)
friction_coef = 0.1
a = 0
t = 0.0
dt = 0.01
pos = []
vel = []
time =[]

while t<=10:
    stretch = startX-h
    spring_force = k*stretch
    a = ((spring_force) + (-w))/m
    v += a*dt
    h += v*dt
    t += dt
    time.append(t)
    pos.append(h)
    vel.append(v)
```

```python
h = 0.5
startX = 0.7
stretch = startX-h
v = 0
k = 5
m = 0.0609
g = 9.8
w = (9.8*m)
air_resistance = 0.1
a = 0
t = 0.0
dt = 0.01
posDamp = []
velDamp = []
time = []
drivForce = []


while t<=10:
    stretch = startX-h
    spring_force = k*stretch
    damping_force = air_resistance*v*v
    driving_force = 1/4*math.sin(3*math.pi*t)
    a = ((spring_force)+(-w) + (driving_force)- np.sign(v)*(damping_force))/m
    v += a*dt
    h += v*dt
    t += dt
    time.append(t)
    drivForce.append(driving_force)
    posDamp.append(h)  
    velDamp.append(v)
```

```python
plt.plot(time, posDamp)
#plt.plot(time, pos)
#plt.plot(time, drivForce)
plt.legend(["posDamp", "pos", "drivForce"])
plt.show()
```
